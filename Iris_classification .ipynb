```python
import tensorflow as tf
def readTrainTestData():
    
    col_norm = ["sepal_length" , "sepal_width" , "petal_length" , "petal_width"]
    X_DATA = dataset[col_norm]
    X_DATA = X_DATA.apply(lambda x:((x-x.min())/(x.max()-x.min())))
   
    lb = LabelEncoder()
    Y_DATA = dataset["species"]
    Y_DATA = dataset["species"].map({"iris-setosa" : 0 , "iris-versicolor" : 1 , "iris-virginica" : 2})

    X_TRAIN , X_TEST , Y_TRAIN , Y_TEST = train_test_split(X_DATA , Y_DATA , test_size = 0.3 , random_state = 101)

    return X_TRAIN , X_TEST , Y_TRAIN , Y_TEST
```

```python
def create_feature_column():
    feat_sepal_length = tf.feature_column.numeric_column("sepal_length")
    feat_sepal_width = tf.feature_column.numeric_column("sepal_width")
    feat_petal_length = tf.feature_column.numeric_column("petal_length")
    feat_petal_width = tf.feature_column.numeric_column("petal_width")

    feature_column = [feat_sepal_length , feat_sepal_width, feat_petal_length , feat_petal_width]
    return feature_column
```

```python
X_TRAIN , X_TEST , Y_TRAIN , Y_TEST = readTrainTestData()
feature_column = create_feature_column()

input_func = tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_TRAIN , y=Y_TRAIN , batch_size = 40 , num_epochs = 1000 , shuffle = True)

eval_input_func = tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_TEST , y=Y_TEST , batch_size = 40 , num_epochs = 1 , shuffle = False)

predict_input_fn =tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_TEST, num_epochs = 1 , shuffle = False)
```

```python
model = tf.estimator.LinearClassifier(feature_columns = feature_column, n_classes = 3)
```

```python
history = model.train(input_fn = input_func , steps = 1000)
```

```python
results = model.evaluate(eval_input_func)
```
